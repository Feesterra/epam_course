```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oo = pd.read_csv('/home/alex/Olympics.csv', skiprows=4)
```

```python
# task10.2 - График числа всех медалей
hh = oo.groupby(['Edition']).agg({'Medal':['count']})
hh
```

```python
plot = hh.plot(kind='bar', title='Number of medals')
```

```python
# спиок стран, кол-ва медалей, первый и последний годы
oo.groupby('NOC').agg({'Medal':['count'], 'Edition':['min', 'max']})
```
