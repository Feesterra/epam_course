```python
import pandas as pd
import numpy as np
```

```python
oo = pd.read_csv('/home/alex/Olympics.csv', skiprows=4)
```

```python
ath_usa = oo[oo.NOC == 'USA']
grouped = ath_usa.groupby(['Edition', 'Athlete', 'Sport', 'Discipline', 'Medal']).size().unstack('Medal', fill_value=0)
```

```python
grouped['Total'] = grouped['Gold'] + grouped['Bronze'] + grouped['Silver']
grouped
```

```python
index = ['Edition', 'Athlete', 'Sport', 'Discipline', 'Bronze', 'Gold', 'Silver', 'Total']
```

```python
gr = grouped.reset_index(col_fill=index)
del gr.columns.name
gr
```

```python
def maximum(gr, column='Total'):
    m = max(gr.Total)
    for i in gr.index:
        if gr.Total[i] == m:
            return gr.loc[i]
maximum(gr)
```

```python
result = gr.groupby('Edition').apply(maximum)
result = result.rename_axis('Idx')
result = result.reset_index(drop=True)
result
```
